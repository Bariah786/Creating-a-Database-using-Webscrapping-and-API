```python
#Creating Function

def City_Weather_info(cities, return_first_df=False, return_second_df=False):

    #API_key from ninja
    API_key= userdata.get('city_API')

    # Initialize lists for storing data across multiple cities
    all_city_data_parameters = []
    all_city_data_cities = []

    for name in cities:
        # Requesting API
        api_url = 'https://api.api-ninjas.com/v1/city?name={}'.format(name)
        response = requests.get(api_url, headers={'X-Api-Key': API_key})

        #COOrdinates of cities
        city_json= response.json()
        lat= city_json[0]['latitude']
        lon= city_json[0]['longitude']

        #API_key for weather api
        api_key= userdata.get('weather_API')
        weather_api= requests.get(f'https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}&units=metric')

        #Create json
        weather_json =  weather_api.json()

        city_id= []
        city_name= []
        city_country= []
        sunrise_time= []
        sunset_time= []
        temperature= []
        min_temp= []
        max_temp= []
        humidity= []
        wind_speed= []
        description= []
        time_of_record= []

        #CITY ID, Name, Country
        cityid= weather_json['city']['id']
        city_id.append(cityid)
        city= city_json[0]['name']
        city_name.append(city)
        country= weather_json['city']['country']
        city_country.append(country)

        #SUNSET

        timestamp= weather_json['city']['sunset']

        # Convert to a readable time format
        sunset_time_dt = datetime.datetime.utcfromtimestamp(timestamp)
        sunset_time.append(sunset_time_dt)

        #SUNRISE

        timestamp= weather_json['city']['sunrise']

        #Convert to a readable format
        sunrise_time_dt= datetime.datetime.utcfromtimestamp(timestamp)
        sunrise_time.append(sunrise_time_dt)

        #Coordinates
        latitude= weather_json['city']['coord']['lat']

        longitude= weather_json['city']['coord']['lon']

        for x in range(len(weather_json['list'])):
            #Temperature
            tem= weather_json['list'][x]['main']['temp']
            temperature.append(tem)

            #Min_Temperature & max_Temperature
            min_t= weather_json['list'][x]['main']['temp_min']
            max_t= weather_json['list'][x]['main']['temp_max']
            min_temp.append(min_t)
            max_temp.append(max_t)

            #Humidity
            hum= weather_json['list'][x]['main']['humidity']
            humidity.append(hum)

            #Wind Speed
            speed= weather_json['list'][x]['wind']['speed']
            wind_speed.append(speed)

            #Description
            desc= weather_json['list'][x]['weather'][0]['description']
            description.append(desc)

            #Time of record
            timestamp= weather_json['list'][x]['dt_txt']

            time_of_record.append(timestamp)

        #Creating DF for weather info
        Parameters= pd.DataFrame({'Name': city_name * len(weather_json['list']),
                        'Temperature': temperature,
                        'Minimum_Temperature': min_temp,
                        'Maximum_Temperature': max_temp,
                        'Humidity': humidity,
                        'Wind_Speed': wind_speed,
                        'Description': description,
                        'Time_of_record':time_of_record}
        )

        #Creating DF for cities
        Cities= pd.DataFrame({'Id':city_id,
                  'Name': city_name,
                  'Country': city_country,
                  'Sunrise': sunrise_time,
                  'Sunset': sunset_time,
                  'Latitude': latitude,
                  'Longitude': longitude})
        # Append the DataFrames for the current city to the lists
        all_city_data_parameters.append(Parameters)
        all_city_data_cities.append(Cities)

    #Concatenate all the dataframes in the list to a single dataframe
    all_city_data_parameters = pd.concat(all_city_data_parameters, ignore_index=True)
    all_city_data_cities = pd.concat(all_city_data_cities, ignore_index=True)


    if return_first_df and return_second_df:
      return all_city_data_parameters, all_city_data_cities  # Return both DataFrames for all cities as a list of tuples
    elif return_first_df:
      return all_city_data_parameters # Return only Parameters (weather data) for all cities
    elif return_second_df:
      return all_city_data_cities  # Return only Cities (city-related data) for all cities
    else:
      return None  # Return None if no DataFrame is requested

```

```python
#Info needed to build the function

city_name= 'berlin'
API_key= userdata.get('weather_city_API')

cities= requests.get(f'http://api.openweathermap.org/data/2.5/forecast?q={city_name}&appid={API_key}')
```

```python
#Creating a function that yeild dataframes with cities and weather info

def Weather_info(city_names, return_first_df=False, return_second_df=False):
    #API key
    API_key= userdata.get('weather_city_API')

    # Initialize lists for storing data across multiple cities
    all_city_data_parameters = []
    all_city_data_cities = []

    # Loop over all cities provided as arguments
    for city_name in city_names:
        # Requesting API
        cities = requests.get(f'http://api.openweathermap.org/data/2.5/forecast?q={city_name}&appid={API_key}&unit=metric')

        # Get the json response
        cities_json = cities.json()

        city_id= []
        city_name_list= []
        city_country= []
        sunrise_time= []
        sunset_time= []
        temperature= []
        min_temp= []
        max_temp= []
        humidity= []
        wind_speed= []
        description= []
        time_of_record= []
        latitude= []
        longitude= []

        #CITY ID, Name, Country
        cityid= cities_json['city']['id']
        city_id.append(cityid)
        city= cities_json['city']['name']
        city_name_list.append(city)
        country= cities_json['city']['country']
        city_country.append(country)

        #SUNSET

        timestamp= cities_json['city']['sunset']

        # Convert to a readable time format
        sunset_time_dt = datetime.datetime.utcfromtimestamp(timestamp)
        sunset_time.append(sunset_time_dt)

        #SUNRISE

        timestamp= cities_json['city']['sunrise']

        #Convert to a readable format
        sunrise_time_dt= datetime.datetime.utcfromtimestamp(timestamp)
        sunrise_time.append(sunrise_time_dt)

        #Coordinates
        lat= cities_json['city']['coord']['lat']
        latitude.append(lat)
        lon= cities_json['city']['coord']['lon']
        longitude.append(lon)

        for x in range(len(cities_json['list'])):
            #Temperature
            tem= cities_json['list'][x]['main']['temp']
            temperature.append(tem)

            #Min_Temperature & max_Temperature
            min_t= cities_json['list'][x]['main']['temp_min']
            max_t= cities_json['list'][x]['main']['temp_max']
            min_temp.append(min_t)
            max_temp.append(max_t)

            #Humidity
            hum= cities_json['list'][x]['main']['humidity']
            humidity.append(hum)

            #Wind Speed
            speed= cities_json['list'][x]['wind']['speed']
            wind_speed.append(speed)

            #Description
            desc= cities_json['list'][x]['weather'][0]['description']
            description.append(desc)

            #Time of record
            timestamp= cities_json['list'][x]['dt_txt']
            time_of_record.append(timestamp)

        #Creating DF for weather info
        Parameters= pd.DataFrame({'Name': [city_name] * len(cities_json['list']),
                                'Temperature': temperature,
                                'Minimum_Temperature': min_temp,
                                'Maximum_Temperature': max_temp,
                                'Humidity': humidity,
                                'Wind_Speed': wind_speed,
                                'Description': description,
                                'Time_of_record':time_of_record}
        )

        #Creating DF for cities
        Cities= pd.DataFrame({'Id':city_id,
                        'Name': city_name,
                        'Country': city_country,
                        'Sunrise': sunrise_time,
                        'Sunset': sunset_time,
                        'Latitude': latitude,
                        'Longitude': longitude})

        # Append the current city's data to the all_city_data list
        all_city_data_parameters.append(Parameters)
        all_city_data_cities.append(Cities)

    #Concatenate all the dataframes in the list to a single dataframe
    all_city_data_parameters = pd.concat(all_city_data_parameters, ignore_index=True)
    all_city_data_cities = pd.concat(all_city_data_cities, ignore_index=True)

    if return_first_df and return_second_df:
        return all_city_data_parameters, all_city_data_cities  # Return both DataFrames for all cities as a list of tuples
    elif return_first_df:
        return all_city_data_parameters # Return only Parameters (weather data) for all cities
    elif return_second_df:
        return all_city_data_cities  # Return only Cities (city-related data) for all cities
    else:
        return None  # Return None if no DataFrame is requested
```
